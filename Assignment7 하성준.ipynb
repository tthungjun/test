**<span style="color:#ffdce0">문제 1번</span>**</br>
</br>
n이 입력될때 첫번째 줄부터 n번째 줄까지 별삼각형을 출력하라</br>
</br>
조건 1. n이 입력되면 첫번째 줄부터 n*2-1, n*2-3 ... 개의 별이 출력된다.</br>
</br>
조건 2. 별은 가운데를 기준으로 대칭이 되어야하고 n이 0~10일때 조건에 맞는 별삼각형이 출력이 되어야한다.</br>
</br>힌트: 출력할 때 print 함수의 파라미터(end=)를 조절할 필요가 있다</br>
</br>
입력 형식:</br>
아래 코드 참조</br>
</br>
출력 형식:</br>
아래 코드 참조
</br>

```python
"""
입력 형식:
0

출력 형식:



입력 형식:
3

출력 형식:
*****
 ***
  *

입력 형식:
5

출력 형식:
********* 
 ******* 
  *****
   *** 
    *

"""
```

```python
#문제 1번

n = int(input())

for i in range(1, n+1):
    print(" "*(i-1), end = '')
    print("*"*(2*(n-i)+1))

```

**<span style="color:#ffdce0">문제 2번</span>**</br>
</br>
입력된 수의 팩토리얼을 출력하라</br>
</br>
조건 1. 정수 n이 입력된다. (0< n < 13)</br>
</br>
조건 2. 팩토리얼을 구하는 함수 fact를 만들고, 만든 fact 함수를 사용하여 팩토리얼을 구해 출력한다.</br>
</br>
입력 형식:</br>
6</br>
</br>
출력 형식:</br>
n! : 720</br>
</br>

```python
#문제 2번
n = int(input())

def fact(n):
    c = 1 
    for a in range(1, n+1):
        c *= a

    return c
    
    
print(f'{n}! :', fact(n))
```

**<span style="color:#ffdce0">문제 4번</span>**</br>
</br>
1부터 입력받은 수까지 범위의 짝수를 리스트에 담아 출력하는 프로그램을 작성하시오.</br>
</br>
조건 1. 빈 리스트 even을 미리 선언한다.</br>
</br>
조건 2. input()를 이용하여 입력을 받고 입력받은 수를 int()를 이용해 정수로 변환하라.</br>
</br>
조건 3. 짝수 값을 리스트 even에 append()를 이용해 담아 리스트를 출력한다.</br>
</br>
입력 형식:</br>
10</br>
</br>
출력 형식:</br>
[2, 4, 6, 8, 10]</br>

```python
#문제 4번

even = list()
a = int(input())
i = 1

while i <= a :
    if i % 2 == 0:
        even.append(i)
    i += 1
    
print(even)
```

**<span style="color:#ffdce0">문제 5번</span>**</br>
</br>
Up & Down 게임을 만들어 보자.</br>
</br>
조건 1. while문을 사용하여 코드를 작성한다. (for문 사용 x)</br>
</br>
조건 2. 사용자가 맞춰야할 값을 미리 answer 변수로 초기화한다.</br>
</br>
조건 3. 정수 n을 입력 받아라. ( 0 < n <= 10 )</br>
</br>
조건 4. 입력한 값보다 answer값이 작으면 Down, 입력한 값이 answer 보다 크면 Up, 입력값과 answer가 같으면 Answer!을 출력한다.</br>
</br>
입력 형식:(answer=4)</br>
5</br>
2</br>
4</br>
</br>
출력 형식:</br>
Down</br>
Up</br>
Answer!</br>

```python
#문제 5번
from random import*
n = int(random()*10)+1
i = 0

while i != n :
    i = int(input())

    if i < n :
        print("Up")
    elif i > n :
        print("Down")
    elif i == n :
        print("Answer!")
```
